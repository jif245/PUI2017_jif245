```python
#Importing packages

import numpy as np
import matplotlib.pyplot as plt

%pylab inline
```

```python
#we set variables that will contain the parameters of the distribution
media=12 #this will the mean 
desvio=10 #this will the standard deviation

np.random.seed(14) #setting a seed for reproducible research

#we graph a point in each loop
for i in range(100):
    random_size=np.random.randint(10,2000) #we create a random number between 10 and 2000 for sample size
    sample=np.random.normal(media,desvio,random_size) #we create a sample with the parameters and random size
    plt.plot(sample.mean(),random_size,"o"); #we plot the mean of that sample

plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Sample Size", fontsize=14)
plt.title('Fig1 .Normal Distribution',weight='bold',fontsize=16)
 
```

```python
#We do the same for the poisson distribution
np.random.seed(14)

#setting lambda
fish=12

for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.poisson(fish,random_size)
    plt.plot(sample.mean(),random_size,"o");

plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Sample Size", fontsize=14)
plt.title('Fig 2. Poisson Distribution',weight='bold',fontsize=16)
```

```python
#Binomial distribution
np.random.seed(14)

#n*p=12, setting the variables of binomial
n=24
p=0.5

for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.binomial(n,p,random_size)
    plt.plot(sample.mean(),random_size,"o");

plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Sample Size", fontsize=14)
plt.title('Fig 3. Binomial Distribution',weight='bold',fontsize=16)
```

```python
#Chi-square distribution
np.random.seed(14)

dfreedom=12

for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.chisquare(dfreedom,random_size)
    plt.plot(sample.mean(),random_size,"o");

plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Sample Size", fontsize=14)
plt.title('Fig 4. Chi-Square',weight='bold',fontsize=16)
```

```python
#Exponential distribution
np.random.seed(14)

scale=12

for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.exponential(scale,random_size)
    plt.plot(sample.mean(),random_size,"o");

plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Sample Size", fontsize=14)
plt.title('Fig 5. Exponential',weight='bold',fontsize=16)
```

```python
np.random.seed(14)

media=scale=fish=12
desvio=10
n=24
p=0.5

#For normal distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.normal(media,desvio,random_size)
    plt.plot(random_size,sample.mean(),"o",color="red");

#For Poisson distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.poisson(fish,random_size)
    plt.plot(random_size,sample.mean(),"o",color="blue");

#For Chi-Square distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.chisquare(dfreedom,random_size)
    plt.plot(random_size,sample.mean(),"o",color="grey");

#For Binomial Distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.binomial(n,p,random_size)
    plt.plot(random_size,sample.mean(),"o",color="green");

#For Exponential Distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.exponential(scale,random_size)
    plt.plot(random_size,sample.mean(),"o",color="black");
    
plt.xlabel("Sample Size", fontsize=14)
plt.ylabel("Sample Mean", fontsize=14)
plt.title('Fig 6. All Distributions',weight='bold',fontsize=16)
```

*Figure 6*. Bigger samples sizes have less deviaton in their mean value.
The bigger the sample, the more similar is to the theoretical value of the mean.


```python
import numpy as np
import scipy.stats as stats
import pylab as pl


np.random.seed(14)

media=scale=fish=dfreedom=12
desvio=10
n=24
p=0.5
meanlist=[]

#For normal distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.normal(media,desvio,random_size)
    meanlist.append(sample.mean())

#For Poisson distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.poisson(fish,random_size)
    meanlist.append(sample.mean())

#For Chi-Square distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.chisquare(dfreedom,random_size)
    meanlist.append(sample.mean())

#For Binomial Distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.binomial(n,p,random_size)
    meanlist.append(sample.mean())

#For Exponential Distribution
for i in range(100):
    random_size=np.random.randint(10,2000)
    sample=np.random.exponential(scale,random_size)
    meanlist.append(sample.mean())

#Fitting to a gaussian model:
sorted_meanlist = sorted(meanlist)  #sorted
fit = stats.norm.pdf(sorted_meanlist, np.mean(sorted_meanlist), np.std(sorted_meanlist))  #fitting model
plt.plot(sorted_meanlist,fit)
plt.hist(sorted_meanlist,normed=True, bins=50)      #use this to draw histogram of your data
   
plt.xlabel("Sample Mean", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title('Fig 7. Distribution of Means',weight='bold',fontsize=16)

# The binomial fit method was obtained from the following site:
#source:https://stackoverflow.com/questions/20011494/plot-normal-distribution-with-matplotlib
```

Fig 7. Central Limit Theroem. The distribution of means has a normal distribution.
